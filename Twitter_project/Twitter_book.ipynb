```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json

from pandas.io.json import json_normalize #package for flattening json in pandas df

# prints whole output
# pd.set_option('display.max_colwidth', -1)
```

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
with open('/Users/matthewmello/bigdata/Twitter_project/data/food/flattened_mcdonalds_tweets.txt', 'w') as wf:
    with open('/Users/matthewmello/bigdata/Twitter_project/data/food/mcdonalds_tweets.txt') as rf:
        for i,line in enumerate(rf):
            json_acceptable_string = line.replace("'", r"\"")
            d = json.loads(json_acceptable_string)
            flat = flatten_json(d)
            json_normalize(flat)
            json.dump(flat, wf)
            wf.write('\n')
```
